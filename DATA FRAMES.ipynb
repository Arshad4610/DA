# Perform data aggregation

```python
a=k.groupby("educational_level").agg({'annual_income' : 'mean','recency' : 'mean','online_purchases' : 'mean','store_purchases' : 'mean'}).reset_index
a
```

```python
a=k.groupby(by="educational_level")["annual_income","recency","store_purchases","online_purchases"].mean().reset_index()
a
```

```python
b=pd.pivot_table(k,values=["sum_of_purchases","recency"],
                 index=["marital_status"],
                        columns=["educational_level"],
                        aggfunc={'sum_of_purchases':np.sum,'recency':np.mean}).reset_index()
b
```
